```python
# Importing Libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset
```

```python
df = pd.read_excel('Online retail.xlsx', header=None)
```

```python
df.columns = ['Items']
```

```python
df
```

```python
# Creating Transaction Matrix
```

```python
# Split the 'Items' column into lists of items
df['Items'] = df['Items'].apply(lambda x: x.split(','))
```

```python
df['Items']
```

```python
# Get a list of unique items
unique_items = sorted(set(item for sublist in df['Items'] for item in sublist))
```

```python
unique_items
```

```python
transaction_item_matrix = pd.DataFrame(columns=unique_items)
```

```python
transaction_item_matrix
```

```python
for _, row in df.iterrows():
    transaction_dict = {item: 1 if item in row['Items'] else 0 for item in unique_items}
    transaction_item_matrix = transaction_item_matrix.append(transaction_dict, ignore_index=True)
```

```python
transaction_item_matrix
```

```python
# Filling NaN value in Transaction Matrix
```

```python
transaction_item_matrix.fillna(0, inplace=True)
```

```python
transaction_item_matrix
```

```python
# Applying Apriori Algorithm to find frequent dataset
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(transaction_item_matrix, min_support=0.03, use_colnames=True)
```

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
```
