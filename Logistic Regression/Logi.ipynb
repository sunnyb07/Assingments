```python
# Doing one hot encoding on Training and Testing data
```

```python
data_train = pd.get_dummies(data=data_train, columns=['Embarked'])
```

```python
data_test = pd.get_dummies(data=data_test, columns=['Embarked'])
```

```python
data_train
```

```python
# Splitting data in X and y
```

```python
X_train = data_train.drop(columns='Survived')
```

```python
X_train
```

```python
y_train = data_train['Survived']
```

```python
X_test = data_test
```

# Creating and Fitting the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=400)
model.fit(X_train, y_train)
```

# Model Evalution

```python
y_pred = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
```

```python
print('Accuracy Score:', accuracy_score(y_train, y_pred))
```

```python
print('ROC Score:', roc_auc_score(y_train, y_pred))
```

```python
# plotting roc curve
```

```python
fpr, tpr, thresholds = roc_curve(y_train, model.predict_proba(X_train)[:,1])
plt.plot(fpr, tpr, label='ROC Curve (area = %0.2f)' % roc_auc_score(y_train, y_pred))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

# Predictting the model

```python
y_pred = model.predict(X_test)
```
