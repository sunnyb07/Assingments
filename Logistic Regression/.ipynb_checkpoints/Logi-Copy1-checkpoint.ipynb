```python
# Here we drop 4 columns Which are not useful i.e. passangerId, Name, Ticket and Cabin
```

```python
# Create the subplots layout with 1 row and 2 columns, set the figure size to (15,8)
f, ax = plt.subplots(1, 2, figsize=(15, 8))

# Plot a pie chart to describe the distribution of the 'Survived' variable
# 1 represents Survived passengers, 0 represents Not Survived passengers
# 'explode' specifies the fraction of the radius with which to offset each wedge
data_train['Survived'].value_counts().plot.pie(ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('Survivors')
ax[0].set_ylabel('')

# Plot a count plot to show the count of Survived and Not Survived passengers based on gender ('Sex')
sns.countplot(data=data_train, x='Sex', hue='Survived', ax=ax[1])
ax[1].set_title('Survivors Count')

# Show the plots
plt.show()
```

```python
# Plotting the count of passengers by class
data_train['Pclass'].value_counts().plot.bar(color=['#CD7F32', '#FFDF00', '#D3D3D3'])
plt.title('Passenger Count by Class')
plt.ylabel('Count')
plt.xlabel('Class')
plt.show()
```
