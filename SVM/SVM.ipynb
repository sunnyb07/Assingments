```python
# Label Encoding class column
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(df['class'])
```

```python
df['class'] = labels
```

```python
df['class']
```

```python
# doing one hot encoding on categorical cols
```

```python
df.info()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
```

```python
categorical_df_encoded = pd.get_dummies(df)
```

```python
categorical_df_encoded
```

```python
# Data Transformation
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df1 = categorical_df_encoded.drop('class', axis=1)
```

```python
scaler.fit(df1)
```

```python
transform = scaler.transform(df1)
```

```python
scaled_df = pd.DataFrame(transform, columns=df1.columns)
```

```python
scaled_df
```

```python
# Splitting data in training and testing
```

```python
X = scaled_df
```

```python
y = categorical_df_encoded['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Creatting and fitting the model
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
# Model Evaluation
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Parameter Tuning
```

```python
# Selecting different kernels for model
```

```python
# Using linear kernel
```

```python
model1 = SVC(kernel='linear')
```

```python
model1.fit(X_train, y_train)
```

```python
y_pred1 = model1.predict(X_test)
```

```python
print(classification_report(y_test, y_pred1))
```

```python
# Using polynolial kernel
```

```python
model2 = SVC(kernel='poly')
```

```python
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_test)
```

```python
print(classification_report(y_test, y_pred2))
```

```python
# Using RBF model
```

```python
model3 = SVC(kernel='rbf')
```

```python
model3.fit(X_train, y_train)
```

```python
y_pred3 = model3.predict(X_test)
```

```python
print(classification_report(y_test, y_pred3))
```
