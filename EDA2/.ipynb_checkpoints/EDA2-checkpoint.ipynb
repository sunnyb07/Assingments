# Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading the dataset

```python
df = pd.read_csv('adult_with_headers.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Replacing '?' values in dataset with NaN
```

```python
df.isnull().sum()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.fillna(method='bfill', inplace=True)
```

```python
df
```

```python
df.info()
```

```python
# Creating 2 new Features
```

```python
# for Age

print(min(df['age']))
print(max(df['age']))
```

```python
# Here Age differ from 17 to 90

# We Create 3 different type of age groups
# for 17 to 25 as 'Young'
# for 25 to 50 as 'Middle age'
# for 50+ as 'Old age'
```

```python
def categorize_age(age):
    if age >= 17 and age <= 25:
        return 'Young'
    elif age > 25 and age <= 50:
        return 'Middle age'
    elif age > 50:
        return 'Old age'
```

```python
df['age'] = df['age'].apply(categorize_age)
```

```python
df['age'].value_counts()
```

```python
# We will do same for hours_per week
```

```python
print(min(df['hours_per_week']))
print(max(df['hours_per_week']))
```

```python
# Here hours_per_week differ from 1 to 99

# We Create 3 different type of age groups
# for <40 as 'Part time'
# for 45 to 55 as 'Full Time'
# for 55+ as 'Over Time'
```

```python
def categorize_hour(hours_per_week):
    if hours_per_week < 40:
        return 'Part Time'
    elif hours_per_week > 15 and hours_per_week <= 55:
        return 'Full Time'
    elif hours_per_week > 55:
        return 'Over Time'
```

```python
df['hours_per_week'] = df['hours_per_week'].apply(categorize_hour)
```

```python
df['hours_per_week'].value_counts()
```

```python
# Doing label Encoding on income column since it is target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
# converting <=50K as 0 and >50K as 1 
```

```python
df['income'] = label_encoder.fit_transform(df['income'])
```

```python
df
```

```python
# Doing one hot encoding on columns whose unique value are less than 5 and Label Encoding on other
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
```

```python
categorical_columns
```

```python
# Initialize empty DataFrame to store encoded data
encoded_df = pd.DataFrame()
```

```python
for col in categorical_columns:

    if df[col].nunique() < 5:
        encoded_col_df = pd.get_dummies(df[col], prefix=col)
    else:
        encoded_col = label_encoder.fit_transform(df[col])
        encoded_col_df = pd.DataFrame(encoded_col, columns=[col+'_encoded'])
    
    encoded_df = pd.concat([encoded_df, encoded_col_df], axis=1)
```

```python
encoded_df
```

```python
df.drop(columns=categorical_columns, inplace=True)
df = pd.concat([df, encoded_df], axis=1)
```
