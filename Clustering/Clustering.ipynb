```python
# data preprocessing
```

```python
# Standarizing the data using MinMax Scaler
```

```python
# Using Minmaxscaler for accuracy result comparison

from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()

minmax_df = minmax.fit_transform(df)
minmax_df.shape
```

```python
minmax_df
```

```python
# Modelling
```

```python
# Creating Kmeans model for clustering
```

```python
from sklearn.cluster import KMeans
```

```python
k_c = KMeans(4,random_state=4) #Keeping cluster as 4
```

```python
k_c.fit_predict(minmax_df)
```

```python
df['Cluster_no'] = pd.DataFrame(k_c.fit_predict(minmax_df),columns=['Cluster_no'])
```

```python
df
```

```python
# Getting best k for cluster
```

```python
wcss = []
for i in range(1,25):
    k_c = KMeans(n_clusters = i,random_state = 42)
    k_c.fit_predict(minmax_df)
    wcss.append(k_c.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,25),wcss)
```

```python
# Here we get elbow at 5 so best cluster will be at 5
```

```python
k_c = KMeans(5,random_state=4)
```

```python
k_c.fit_predict(minmax_df)
```

```python
df['Cluster_no'] = pd.DataFrame(k_c.fit_predict(minmax_df),columns=['Cluster_no'])
```

```python
df
```

```python
sns.scatterplot(x = df['Balance'],y = df['Bonus_miles'],hue=df['Cluster_no'],palette='Set1')
```

```python
# Using DB Scan model
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=2, min_samples=10)
dbscan.fit(minmax_df)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
clustered = pd.concat([df,cl],axis=1)
```

```python
clustered
```

```python
sns.scatterplot(clustered['Balance'],clustered['Bonus_miles'],hue=clustered['cluster'],palette='Set1')
```

```python
# Using Agglomerative Clustering
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc1 = AgglomerativeClustering(n_clusters=5, linkage='ward')
```

```python
hc1.fit(minmax_df)
```

```python
hc1_pred = hc1.fit_predict(minmax_df)
```

```python
hc1_pred
```

```python
df['cluster_no'] = pd.DataFrame(hc1_pred,columns=['cluster_no'])
```

```python
df
```

```python
sns.scatterplot(x = df['Balance'],y = df['Bonus_miles'],hue=df['cluster_no'],palette='Set1')
```
